Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df['price_per_sqft'] = df['price'] / df['sqft_living']
df.head()
```

Изучим зависисмость цены квадратного фута от жилой площади дома

```python
pivot_table_sqft_living = df.pivot_table(index = 'sqft_living', values = 'price_per_sqft', aggfunc = ['mean', 'count', 'median'])
pivot_table_sqft_living.columns = ['mean', 'count', 'median']
pivot_table_sqft_living.plot(y = 'median', style = 'o')

pivot_table_sqft_living.sort_values('median', ascending = False)

df['sqft_living'].corr(df['price_per_sqft'])
```

Вывод: согласно корреляции Пирсона - 0.09 - зависимость очень слабая. 

Изучим зависимость цены квадратного фута от количества спален

```python
pivot_table_bedrooms = df.pivot_table(index = 'bedrooms', values = 'price_per_sqft', aggfunc = ['mean', 'count', 'median'])
pivot_table_bedrooms.columns = ['mean', 'count', 'median']
pivot_table_bedrooms.query('count > 50').plot(y = 'median')

pivot_table_bedrooms.query('count > 50').sort_values('median', ascending = False)

df['bedrooms'].corr(df['price_per_sqft'])
```

Вывод: в домах с количеством спален 3 и более стоимость квадратного фута значительно меньше
